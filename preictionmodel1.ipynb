```python
import os
import cv2
import numpy as np
import shap
from lime import lime_image
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.imagenet_utils import preprocess_input
from tensorflow.keras.preprocessing.image import img_to_array
from skimage.segmentation import mark_boundaries
import matplotlib.pyplot as plt

# Load the trained model
model = load_model('brain_tumour_cnn_model(sample).h5')

# Define image size and data generator parameters
img_height, img_width = 128, 128
batch_size = 32

# Data augmentation & rescaling
datagen = ImageDataGenerator(
    rescale=1.0/255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2  # **Splitting 20% data for validation**
)

# Training Data Generator (80% of data)
train_generator = datagen.flow_from_directory(
    directory="OCT2017_subset/train",  # **Use only train directory**
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode="categorical",
    subset="training"  # **80% of data for training**
)

# Validation Data Generator (20% of data)
val_generator = datagen.flow_from_directory(
    directory="OCT2017_subset/train",  # **Same directory**
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode="categorical",
    subset="validation"  # **20% of data for validation**
)

# Get class labels from training generator
class_labels = list(train_generator.class_indices.keys())
print(f"Class labels: {class_labels}")

# Function to preprocess a single image
def preprocess_image(image_path):
    img = cv2.imread(image_path)
    if img is None:
        raise ValueError(f"Could not read image from {image_path}. Check the file path.")
    
    img = cv2.resize(img, (img_height, img_width))
    img = img_to_array(img)
    img = img / 255.0  # Match with ImageDataGenerator rescaling
    return np.expand_dims(img, axis=0)  # Add batch dimension
# Function to predict a single image

def predict_image(image_path):
    image = preprocess_image(image_path)
    prediction = model.predict(image)
    predicted_class = np.argmax(prediction, axis=1)[0]
    confidence = prediction[0][predicted_class]
    return class_labels[predicted_class], confidence, prediction

# Example prediction
image_path = 'OCT2017_subset/train/DRUSEN/DRUSEN-142234-5.jpeg'  # Replace with your image path
try:
    predicted_class, confidence, prediction = predict_image(image_path)
    print(f"Predicted Class: {predicted_class}, Confidence: {confidence * 100:.2f}%")
    print("Model Prediction Distribution:", prediction)
except ValueError as e:
    print(e)

# SHAP analysis for interpretability
image_shape = (img_height, img_width, 3)
masker_blur = shap.maskers.Image("blur(32,32)", image_shape)

# Prediction function for SHAP
def predict_fn(images):
    return model.predict(images / 255.0)  # Ensure normalization consistency


# Initialize SHAP explainer
explainer = shap.Explainer(predict_fn, masker_blur, output_names=class_labels)

# Generate SHAP values
try:
    single_image = preprocess_image(image_path)
    shap_values = explainer(single_image, max_evals=5000, batch_size=1)

    # Plot SHAP values
    shap.image_plot(shap_values, single_image)
except ValueError as e:
    print(e)

# LIME analysis for interpretability
def lime_explanation(image_path):
    lime_explainer = lime_image.LimeImageExplainer()

    # Prediction function for LIME
    def predict_fn_for_lime(images):
        return model.predict(images / 255.0)  # Normalize same as training


    img = cv2.imread(image_path)
    if img is None:
        raise ValueError(f"Could not read image from {image_path}. Please check the file path.")
    img = cv2.resize(img, (img_height, img_width))
    explanation = lime_explainer.explain_instance(
        img.astype(np.double),  # LIME requires double precision
        predict_fn_for_lime,
        top_labels=3,
        hide_color=0,
        num_samples=1000
    )

    temp, mask = explanation.get_image_and_mask(
        explanation.top_labels[0],
        positive_only=True,
        num_features=5,
        hide_rest=False
    )

    plt.figure(figsize=(8, 8))
    plt.imshow(mark_boundaries(temp / 255.0, mask))
    plt.title(f"LIME Explanation for {class_labels[explanation.top_labels[0]]}")
    plt.axis('off')
    plt.show()

# Call LIME explanation
try:
    lime_explanation(image_path)
except ValueError as e:
    print(e)
```
